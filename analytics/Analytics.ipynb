```python
frame["hour_of_day"] = frame.Zeit.dt.hour
frame["day_of_week"] = frame.Zeit.dt.dayofweek
frame["month"] = frame.Zeit.dt.month
```

```python
import sklearn
import sklearn.preprocessing
import sklearn.ensemble
import sklearn.metrics
import sklearn.model_selection
import sklearn.utils
```

```python
le = sklearn.preprocessing.LabelEncoder()
frame["station_label"] = le.fit_transform(frame['Stationscode'])
frame["target"] = frame["Messwert (in µg/m³)"]
```

```python
frame.head()
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(frame[["hour_of_day", "day_of_week", "station_label", "month"]], frame["target"], test_size=0.25, random_state=42, shuffle=True)
```

```python
X_train.shape, y_train.shape
```

```python
X_train_samp, y_train_samp = sklearn.utils.shuffle(X_train, y_train, n_samples=50000)
```

```python
params = {'n_estimators': [350], 'max_depth': [7], 'min_samples_split': [20],
          'learning_rate': [.1], 'loss': ['ls']}
clf = sklearn.ensemble.GradientBoostingRegressor()
gv = sklearn.model_selection.GridSearchCV(clf, params, scoring='r2', verbose=3, n_jobs=3, cv=10)
```

```python
gv.fit(X_train_samp, y_train_samp)
```

```python
gv.best_estimator_
```

```python
gv.best_estimator_.feature_importances_
```

```python
pred_frame = pd.DataFrame(y_test, columns=["target"])
pred_frame["pred"] = gv.predict(X_test)
sklearn.metrics.r2_score(pred_frame["target"], pred_frame["pred"])
```

```python
ret = sklearn.ensemble.partial_dependence.partial_dependence(gv.best_estimator_,[1], X=frame[["hour_of_day", "day_of_week", "station_label", "month"]], percentiles=(0.05, 0.95), grid_resolution=50)
days = pd.DataFrame([ret[0][0],ret[1][0]]).T
days.columns = ["vals", "day"]
ax = days.plot(x= "day", y="vals",kind="scatter", legend=False, s=100, rot=-45)
days_list = ['', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
ax.set_xticklabels(days_list)
ax.set_ylabel("")
```

```python
ret = sklearn.ensemble.partial_dependence.partial_dependence(gv.best_estimator_,[0], X=frame[["hour_of_day", "day_of_week", "station_label", "month"]], percentiles=(0.05, 0.95), grid_resolution=50)
hours = pd.DataFrame([ret[0][0],ret[1][0]]).T
hours.columns = ["vals", "hours"]
hours.set_index("hours", inplace=True)
hours.plot(legend=False, lw=5, c="green")
```

```python
ret = sklearn.ensemble.partial_dependence.partial_dependence(gv.best_estimator_,[2], X=frame[["hour_of_day", "day_of_week", "station_label", "month"]], percentiles=(0.00, 1.), grid_resolution=411)
stations = pd.DataFrame([ret[0][0],ret[1][0]]).T
stations.columns = ["vals", "station"]
stations.set_index("station", inplace=True)
stations.plot(figsize=(10,6),legend=False)
```

```python
stations.vals.argmax()
top_frame = frame[frame.station_label==stations.vals.argmax()]
top_frame.set_index("Zeit", inplace=True)
```

```python

top_frame["pred"] = gv.predict(top_frame[["hour_of_day", "day_of_week", "station_label", "month"]])
top_frame.head()
```
